```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re 

def insta_searching(word):  #wordë¼ëŠ” ë§¤ê°œë³€ìˆ˜ë¥¼ ë°›ëŠ” insta_searching ì´ë¼ëŠ” í•¨ìˆ˜ ìƒì„±
    url = 'https://www.instagram.com/explore/tags/' + word
    return url
def select_first(driver):
    first = driver.find_element_by_css_selector('div._9AhH0') 
    #find_element_by_css_selector í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ ìš”ì†Œ ì°¾ê¸°
    first.click()
    time.sleep(3) #ë¡œë”©ì„ ìœ„í•´ 3ì´ˆ ëŒ€ê¸°
#ë³¸ë¬¸ ë‚´ìš©, ì‘ì„± ì¼ì‹œ, ìœ„ì¹˜ ì •ë³´ ë° í•´ì‹œíƒœê·¸(#) ì¶”ì¶œ
def get_content(driver):
    # 1. í˜„ì¬ í˜ì´ì§€ì˜ HTML ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')    
    # 2. ë³¸ë¬¸ ë‚´ìš© ê°€ì ¸ì˜¤ê¸°
    try:  			#ì—¬ëŸ¬ íƒœê·¸ì¤‘ ì²«ë²ˆì§¸([0]) íƒœê·¸ë¥¼ ì„ íƒ  
        content = soup.select('div.C4VMK > span')[0].text 
        			#ì²« ê²Œì‹œê¸€ ë³¸ë¬¸ ë‚´ìš©ì´ <div class="C4VMK"> ì„ì„ ì•Œ ìˆ˜ ìˆë‹¤.
                                #íƒœê·¸ëª…ì´ div, classëª…ì´ C4VMKì¸ íƒœê·¸ ì•„ë˜ì— ìˆëŠ” span íƒœê·¸ë¥¼ ëª¨ë‘ ì„ íƒ.
    except:
        content = ' ' 
    # 3. ë³¸ë¬¸ ë‚´ìš©ì—ì„œ í•´ì‹œíƒœê·¸ ê°€ì ¸ì˜¤ê¸°(ì •ê·œí‘œí˜„ì‹ í™œìš©)
    tags = re.findall(r'#[^\s#,\\]+', content) # content ë³€ìˆ˜ì˜ ë³¸ë¬¸ ë‚´ìš© ì¤‘ #ìœ¼ë¡œ ì‹œì‘í•˜ë©°, #ë’¤ì— ì—°ì†ëœ ë¬¸ì(ê³µë°±ì´ë‚˜ #, \ ê¸°í˜¸ê°€ ì•„ë‹Œ ê²½ìš°)ë¥¼ ëª¨ë‘ ì°¾ì•„ tags ë³€ìˆ˜ì— ì €ì¥
    # 4. ì‘ì„± ì¼ì ê°€ì ¸ì˜¤ê¸°
    try:
        date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10] #ì•ì—ì„œë¶€í„° 10ìë¦¬ ê¸€ì
    except:
        date = ''
    # 5. ì¢‹ì•„ìš” ìˆ˜ ê°€ì ¸ì˜¤ê¸°
    try:
        like = soup.select('div.Nm9Fw > button')[0].text[4:-1] 
    except:
        like = 0
    # 6. ìœ„ì¹˜ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    try:
        place = soup.select('div.JF9hh')[0].text
    except:
        place = ''
    # 7. ìˆ˜ì§‘í•œ ì •ë³´ ì €ì¥í•˜ê¸°
    data = [content, date, like, place, tags]
    return data
        

def move_next(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow') 
    right.click()
    time.sleep(3)

#1. í¬ë¡¬ìœ¼ë¡œ ì¸ìŠ¤íƒ€ê·¸ë¨ - 'ì‚¬ë‹¹ë§›ì§‘' ê²€ìƒ‰
driver = webdriver.Chrome("./chromedriver.exe")
word = 'ë¬¸ì¬ì¸'
url = insta_searching(word)
driver.get(url) 
time.sleep(4) 
#2. ë¡œê·¸ì¸ í•˜ê¸°
login_section = '//*[@id="loginForm"]/div'
driver.find_element_by_xpath(login_section).click()
time.sleep(3) 
elem_login = driver.find_element_by_name("username")
elem_login.clear()
elem_login.send_keys('soohi0@naver.com') 
elem_login = driver.find_element_by_name('password')
elem_login.clear()
elem_login.send_keys('heyhaewon12!') 
time.sleep(1) 
xpath = '//*[@id="loginForm"]/div/div[3]/button/div'
driver.find_element_by_xpath(xpath).click() 
time.sleep(4) 
xpath1 = """//*[@id="react-root"]/section/main/div/div/div/div/button"""
driver.find_element_by_xpath(xpath1).click()
time.sleep(4) 
#3. ê²€ìƒ‰í˜ì´ì§€ ì ‘ì†í•˜ê¸°
driver.get(url)
time.sleep(4) 
#4. ì²«ë²ˆì§¸ ê²Œì‹œê¸€ ì—´ê¸°
select_first(driver) 
#5. ë¹„ì–´ìˆëŠ” ë³€ìˆ˜(results) ë§Œë“¤ê¸°
results = [] 
#ì—¬ëŸ¬ ê²Œì‹œë¬¼ í¬ë¡¤ë§í•˜ê¸°
target = 2 #í¬ë¡¤ë§í•  ê²Œì‹œë¬¼ ìˆ˜
for i in range(target):
    data = get_content(driver) #ê²Œì‹œë¬¼ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    results.append(data)
    move_next(driver)    
print(results[:2])
#import pandas as pd
from openpyxl import Workbook

write_wb = Workbook()
# # ì´ë¦„ì´ ìˆëŠ” ì‹œíŠ¸ë¥¼ ìƒì„±
# write_ws = write_wb.create_sheet('data1')


#raw_total = pd.read_excel("C:\\Users\\soohi\\Desktop\\real_pub_opinion\\real_pub_opinion.xlsx")
#raw_total.head()
# Sheet1ì—ë‹¤ ì…ë ¥
write_ws = write_wb.active
write_ws['A1'] = 'Content'
write_ws['B1'] = 'Date'
write_ws['C1'] = 'Like'
write_ws['D1'] = 'Place'
write_ws['E1'] = 'Tags'

for j in range(target):
    write_ws.append(results[j]);
    
write_wb.save("C:\\Users\\soohi\\Desktop\\real_pub_opinion\\real_pub_opinion.xlsx")



```
