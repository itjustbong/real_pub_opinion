```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re 

def insta_searching(word):  #wordвЮ╝віћ вДцЖ░юв│ђВѕўвЦ╝ в░Џвіћ insta_searching ВЮ┤вЮ╝віћ ьЋеВѕў ВЃЮВё▒
    url = 'https://www.instagram.com/explore/tags/' + word
    return url
def select_first(driver):
    first = driver.find_element_by_css_selector('div._9AhH0') 
    #find_element_by_css_selector ьЋеВѕўвЦ╝ ВѓгВџЕьЋ┤ ВџћВєї В░ЙЖИ░
    first.click()
    time.sleep(3) #вАювћЕВЮё ВюёьЋ┤ 3В┤ѕ вїђЖИ░
#в│ИвгИ вѓ┤ВџЕ, ВъЉВё▒ ВЮ╝ВІю, ВюёВ╣ў ВаЋв│┤ в░Ј ьЋ┤ВІюьЃюЖиИ(#) ВХћВХю
def get_content(driver):
    # 1. ьўёВъг ьјўВЮ┤ВДђВЮў HTML ВаЋв│┤ Ж░ђВаИВўцЖИ░
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')    
    # 2. в│ИвгИ вѓ┤ВџЕ Ж░ђВаИВўцЖИ░
    try:  			#ВЌгвЪг ьЃюЖиИВцЉ В▓Фв▓ѕВДИ([0]) ьЃюЖиИвЦ╝ ВёаьЃЮ  
        content = soup.select('div.C4VMK > span')[0].text 
        			#В▓Ф Ж▓їВІюЖИђ в│ИвгИ вѓ┤ВџЕВЮ┤ <div class="C4VMK"> ВъёВЮё ВЋї Вѕў ВъѕвІц.
                                #ьЃюЖиИвфЁВЮ┤ div, classвфЁВЮ┤ C4VMKВЮИ ьЃюЖиИ ВЋёвъўВЌљ Въѕвіћ span ьЃюЖиИвЦ╝ вфевЉљ ВёаьЃЮ.
    except:
        content = ' ' 
    # 3. в│ИвгИ вѓ┤ВџЕВЌљВёю ьЋ┤ВІюьЃюЖиИ Ж░ђВаИВўцЖИ░(ВаЋЖиюьЉюьўёВІЮ ьЎюВџЕ)
    tags = re.findall(r'#[^\s#,\\]+', content) # content в│ђВѕўВЮў в│ИвгИ вѓ┤ВџЕ ВцЉ #Вю╝вАю ВІюВъЉьЋўвЕ░, #вњцВЌљ ВЌ░ВєЇвљю вгИВъљ(Ж│хв░▒ВЮ┤вѓў #, \ ЖИ░ьўИЖ░ђ ВЋёвІї Ж▓йВџ░)вЦ╝ вфевЉљ В░ЙВЋё tags в│ђВѕўВЌљ ВађВъЦ
    # 4. ВъЉВё▒ ВЮ╝Въљ Ж░ђВаИВўцЖИ░
    try:
        date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10] #ВЋъВЌљВёювХђьё░ 10Въљвдг ЖИђВъљ
    except:
        date = ''
    # 5. ВбІВЋёВџћ Вѕў Ж░ђВаИВўцЖИ░
    try:
        like = soup.select('div.Nm9Fw > button')[0].text[4:-1] 
    except:
        like = 0
    # 6. ВюёВ╣ў ВаЋв│┤ Ж░ђВаИВўцЖИ░
    try:
        place = soup.select('div.JF9hh')[0].text
    except:
        place = ''
    # 7. ВѕўВДЉьЋю ВаЋв│┤ ВађВъЦьЋўЖИ░
    data = [content, date, like, place, tags]
    return data
        

def move_next(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow') 
    right.click()
    time.sleep(3)

#1. ьЂгвАгВю╝вАю ВЮИВіцьЃђЖиИвъе - 'ВѓгвІ╣вДЏВДЉ' Ж▓ђВЃЅ
driver = webdriver.Chrome("./chromedriver.exe")
word = 'вгИВъгВЮИ'
url = insta_searching(word)
driver.get(url) 
time.sleep(4) 
#2. вАюЖиИВЮИ ьЋўЖИ░
login_section = '//*[@id="loginForm"]/div'
driver.find_element_by_xpath(login_section).click()
time.sleep(3) 
elem_login = driver.find_element_by_name("username")
elem_login.clear()
elem_login.send_keys('soohi0@naver.com') 
elem_login = driver.find_element_by_name('password')
elem_login.clear()
elem_login.send_keys('heyhaewon12!') 
time.sleep(1) 
xpath = '//*[@id="loginForm"]/div/div[3]/button/div'
driver.find_element_by_xpath(xpath).click() 
time.sleep(4) 
xpath1 = """//*[@id="react-root"]/section/main/div/div/div/div/button"""
driver.find_element_by_xpath(xpath1).click()
time.sleep(4) 
#3. Ж▓ђВЃЅьјўВЮ┤ВДђ ВаЉВєЇьЋўЖИ░
driver.get(url)
time.sleep(4) 
#4. В▓Фв▓ѕВДИ Ж▓їВІюЖИђ ВЌ┤ЖИ░
select_first(driver) 
#5. в╣ёВќ┤Въѕвіћ в│ђВѕў(results) вДївЊцЖИ░
results = [] 
#ВЌгвЪг Ж▓їВІювг╝ ьЂгвАцвДЂьЋўЖИ░
target = 2 #ьЂгвАцвДЂьЋа Ж▓їВІювг╝ Вѕў
for i in range(target):
    data = get_content(driver) #Ж▓їВІювг╝ ВаЋв│┤ Ж░ђВаИВўцЖИ░
    results.append(data)
    move_next(driver)    
print(results[:2])
#import pandas as pd
from openpyxl import Workbook

write_wb = Workbook()
# # ВЮ┤вдёВЮ┤ Въѕвіћ ВІюьіИвЦ╝ ВЃЮВё▒
# write_ws = write_wb.create_sheet('data1')


#raw_total = pd.read_excel("C:\\Users\\soohi\\Desktop\\real_pub_opinion\\real_pub_opinion.xlsx")
#raw_total.head()
# Sheet1ВЌљвІц ВъЁваЦ
write_ws = write_wb.active
write_ws['A1'] = 'Content'
write_ws['B1'] = 'Date'
write_ws['C1'] = 'Like'
write_ws['D1'] = 'Place'
write_ws['E1'] = 'Tags'

for j in range(target):
    write_ws.append(results[j]);
    
write_wb.save("C:\\Users\\soohi\\Desktop\\real_pub_opinion\\real_pub_opinion.xlsx")



```
